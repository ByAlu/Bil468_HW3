```python
import KTreeModel

tree = KTreeModel.KTree()
```

```python
import pandas as pd
import math
import time 

csv_file = "results.csv"
df = pd.DataFrame(columns=["k","d","time","accuracy", "precision", "recall", "f1","confusion"])
df.to_csv(csv_file, index=False)

for k in range(2,21):
    data_count = len(tree.train_x)
    dmax = int(math.log(data_count, k))
    for d in range(2, dmax+1):
        start = time.time()
        tree.create_tree(k,d)
        accuracy, precision, recall, f1, cm = tree.validate_frames()
        elapsed = time.time() - start
        df = pd.read_csv(csv_file)
        df.loc[len(df)] = [k, d, elapsed, accuracy, precision, recall, f1, cm]
        df.to_csv(csv_file,index=False)
```
